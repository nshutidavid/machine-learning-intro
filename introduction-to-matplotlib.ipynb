## matplotlib example workflow

```python
# 0. import matplotlib an get it ready to ploting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. prepare data
x = [1,2,3,4]
y = [11,15,10,8]
# 2. set up plot
fig,ax= plt.subplots(figsize=(5,5)) # width, height
# figsize is an optional parameter used in Matplotlib to define the dimensions of a figure. It specifies the width and height of the figure in inches.

# 3. plot the data
ax.plot(x,y)

#4. custom plot
ax.set(title="Simple axis",
       xlabel="x-axis",
       ylabel="y-axis")
# 5. Save & show (save the whole  figure)
fig.savefig("sample-plot.png")
```

## Pyplot vs matplotlib oo method
- when plotting something quickly, it is okay to use pyplot method
- when plotting something advanced use the OO method

# OO Method

```python
heart_disease
```

```python
# patients that are over 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
# Create a scatter plot for age vs. cholesterol, colored by the target column.
#pyplot method
over_50.plot(kind ="scatter",
                  x="age",
                  y="chol",
                  c="target");
```

```python
## 00 method mixed with pyplot
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             colormap="viridis",
             ax = ax);
```

```python
## xlim
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax = ax);
ax.set_xlim([45,100]);
```

```python
over_50.target.values
```

```python
#OO method from scratch
fig,ax =plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
#customize the plot
ax.set(title="Heart disease and cholesterol level",
      xlabel="Age",
      ylabel="Cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle=("--"));
```

```python
over_50.head()
```

```python
# subplot of age,chol thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target'])
#customize ax0
ax0.set(title="Heart disease and cholesterol level",
        xlabel="Age",
        ylabel="Cholesterol")
#add a legend to axo
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a mean line
ax0.axhline(over_50["chol"].mean(),linestyle=("--"))

#Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#customize ax1
ax1.set(title="Thalach level",
        xlabel="Age",
        ylabel="Thalach");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle=("--"));

#Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold",color="darkblue");
```

```python
# sharing the X

# subplot of age,chol thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target'])
#customize ax0
ax0.set(title="Heart disease and cholesterol level",
               ylabel="Cholesterol")
#add a legend to axo
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a mean line
ax0.axhline(over_50["chol"].mean(),linestyle=("--"))

#Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
#customize ax1
ax1.set(title="Thalach level",
        xlabel="Age",
        ylabel="Thalach");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle=("--"));

#Add a title to the figure
fig.suptitle("Heart disease analysis",fontsize=16,fontweight="bold",color="darkred");
```

# Changing Styles Within an Existing Style

```python
# using Cmap
#set the style
plt.style.use("seaborn-v0_8-whitegrid")

#OO method from scratch
fig,ax =plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                    cmap="winter") #this changes the color scheme
#customize the plot
ax.set(title="Heart disease and cholesterol level",
      xlabel="Age",
      ylabel="Cholesterol")

#add a legend
ax.legend(*scatter.legend_elements(),title="Target")

#add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle=("--"));
```

```python
# customizing the y and x limitations using xlim and ylim
# subplot of age,chol thalach
fig,(ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                           sharex=True)
# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50['target'],
                   cmap="winter")
#customize ax0
ax0.set(title="Heart disease and cholesterol level",
               ylabel="Cholesterol")

#change the xo x axis limit
ax0.set_xlim([50,80])

#add a legend to axo
ax0.legend(*scatter.legend_elements(),title="Target")

#Add a mean line
ax0.axhline(over_50["chol"].mean(),linestyle=("--"),color="red")

#Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                   cmap="winter")
#customize ax1
ax1.set(title="Thalach level",
        xlabel="Age",
        ylabel="Thalach");
#change the x1 x and y limit
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a mean line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle=("--"),color="red");

#Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold",color="darkblue");
```
