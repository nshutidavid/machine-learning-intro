```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#import data
students = pd.read_csv("students.csv")
students[:7]
```

```python
# plot age over grade
students.plot(x="Age",y="Grade");
```

```python
students[:3]
```

```python
# Plotting the average score in each subject for all students
subjects = ["Math","Science","English","History","Art","Physical_Education","Computer_Science"]
average_scores = students[subjects].mean()
average_scores.plot(kind="bar");
```

```python
# Calculate the total score for each student
students["Total_Score"] = students[subjects].sum(axis=1)
students.head()
```

```python
# Calculate the percentage
# Assuming each subject is out of 100, so the maximum total score is 700
students["Percentage"] = (students['Total_Score'] / (len(subjects) * 100)) * 100
students
```

```python
# Create a new column 'Above_80' where students with percentage > 80% are assigned 1, else 0
students['Above_90'] = np.where(students['Percentage'] > 90, 1, 0)
students.head()
```

```python
students = students.drop("Mean",axis=1)
```

```python
students.head()
```

```python
students["Grade"].plot.hist(bins=5)
```

```python
students.head()
```

```python
students.plot.hist(subplots =True);
```

```python
#OO method
students.plot(kind ="scatter",
              x="Age",
              y="Percentage",
              c="Above_90",              
             colormap="summer");
              
```

```python
## 00 method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
students.plot(kind="scatter",
              x="Grade",
              y="Percentage",
              c="Above_90",
              colormap="viridis",
              ax = ax)


ax.set_xlim([9,13])
ax.set_ylim([70,100]);
```

```python
#OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
stu_scatter = ax.scatter(x=students["Age"],
                         y=students["Percentage"],
                         c=students["Above_90"])
ax.set(title="Students grades analysis",
       xlabel="Age of students",
       ylabel="Percentage of students")
#add a legend
ax.legend(*stu_scatter.legend_elements(),title="Above 90%")

#add a horizontal line
ax.axhline(students["Percentage"].mean(),linestyle=("--"));

#Add a title to the figure
fig.suptitle("Students annually analysis",fontsize=17,fontweight="bold",color="darkblue");
```
