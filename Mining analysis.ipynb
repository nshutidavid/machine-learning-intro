```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#import data as dataframe
data = pd.read_csv("mining.csv")
data.head()
```

```python
data["Profit"]=data["Profit"].astype(int)
```

```python
# Calculate total production, revenue, expenses, and profit per mine
mine_summary = data.groupby('Mine').agg({
    'Production_Tons': 'sum',
    'Revenue': 'sum',
    'Expenses': 'sum',
    'Profit': 'sum'
}).reset_index()
```

```python
mine_summary
```

```python
mine_summary.plot(x="Mine",y="Profit",kind="bar");
```

```python
mine_summary
```

```python
mine_summary.plot(x="Mine",y="Expenses",kind="bar");
```

```python
# Plotting production performance over years
plt.figure(figsize=(12, 6))
for mine in data['Mine'].unique():
    mine_data = data[data['Mine'] == mine]
    plt.plot(mine_data['Year'], mine_data['Production_Tons'], label=mine)

plt.title('Production Performance Over Years')
plt.xlabel('Year')
plt.ylabel('Production (Tons)')
plt.legend()
plt.show()

# Analyzing profit margins
data['Profit_Margin'] = (data['Profit'] / data['Revenue']) * 100

plt.figure(figsize=(12, 6))
for mine in data['Mine'].unique():
    mine_data = data[data['Mine'] == mine]
    plt.plot(mine_data['Year'], mine_data['Profit_Margin'], label=mine)

plt.title('Profit Margin Over Years')
plt.xlabel('Year')
plt.ylabel('Profit Margin (%)')
plt.legend()
plt.show()
```

```python
# Create a new column 'Above_40' where profit-margin with  > 40 are assigned 1, else 0
data['Above_40'] = np.where(data['Profit_Margin'] > 40.000000, 1, 0)
data.head()
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
# subplot of year,profit,production tons ,revenue,expenses
fig,((ax1,ax2))=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10));
                                       
# add data to ax1
scatter=ax1.scatter(x=data["Year"],
                    y=data["Profit"],
                    c=data["Above_40"],
                   cmap="winter");
#customize ax1
ax1.set(title="Mining Profit analysis",
        xlabel="Year",
        ylabel="Profit in Millions");
#add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Profit Margin");
#Add a mean line
ax1.axhline(data["Profit"].mean(),linestyle=("--"));
#change the x1 x axis limit
ax1.set_xlim([2017,2023])

#add data to ax2
scatter=ax2.scatter(x=data["Ore_Type"],
                    y=data["Revenue"],
                    c=data["Above_40"],
                    cmap="winter");
#customize ax2
ax2.set(title="Mining Production Tons analysis",
        xlabel="Ore_Type",
        ylabel="Revenue in Millions")
#add a legend to ax2
ax2.legend(*scatter.legend_elements(),title="Profit Margin")

#add a mean line
ax2.axhline(data["Revenue"].mean(),linestyle=("--"))
#change the x1 x and y limit

#Add a title to the figure
fig.suptitle("Mining Production Analysis",fontsize=16,fontweight="bold",color="darkred");
```
